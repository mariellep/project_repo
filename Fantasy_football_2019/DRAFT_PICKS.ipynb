```python
#Make a function to see the remaining free agents
def list_free_agents(league,pos,score_dict):
    dict_keys=set(score_dict.keys())
    rankings=pd.read_csv('{:}_2018_stats.csv'.format(pos))
    curr_free_agents=league.free_agents(position=pos)
    curr_list=[p.name for p in curr_free_agents]
    p_stats=rankings[rankings['Name'].isin(curr_list)].sort_values(['Ranking'])
    p_stats['pts']=0
    p_cols=set(p_stats.columns)
    common_cols=dict_keys.intersection(p_cols)
    for k in common_cols:
        mult_val=score_dict[k]
        sel_col=p_stats[k].fillna(0)
        p_stats['pts']=p_stats['pts']+sel_col*mult_val
    p_full=p_stats.head(10).dropna(axis=1,how='all')
    return(p_full[['Name','Ranking','Team','pts']],p_full)
    
```

```python
#SNAKE DRAFT!!
#ROUND 1: RB Pick 
#Position 8
#Before my picks
print(league.draft)
#Look at available top RBs
p_abbrev,p_full=list_free_agents(league,'RB',score_dict)
p_abbrev.sort_values(['pts'],ascending=False)
```
